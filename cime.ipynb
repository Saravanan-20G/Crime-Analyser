```python
import pandas as pd
```

```python
df=pd.read_csv('D:\\Guvi\\Crime\\Crimes_-_2001_to_Present.csv')
```

```python
# Handle missing values and data types
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['Hour'] = df['Date'].dt.hour
df.dropna(subset=['Latitude', 'Longitude'], inplace=True)
```

Temporal Analysis:

```python
import matplotlib.pyplot as plt

# Annual Trends
annual_trends = df.groupby('Year').size()
plt.figure(figsize=(10, 5))
annual_trends.plot()
plt.title('Annual Crime Trends')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.show()

# Monthly Trends
monthly_trends = df.groupby(['Year', 'Month']).size().unstack(fill_value=0)
plt.figure(figsize=(15, 7))
monthly_trends.plot()
plt.title('Monthly Crime Trends')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.show()

# Peak Crime Hours
peak_hours = df.groupby('Hour').size()
plt.figure(figsize=(10, 5))
peak_hours.plot(kind='bar')
plt.title('Crimes by Hour of the Day')
plt.xlabel('Hour')
plt.ylabel('Number of Crimes')
plt.show()
```

Geospatial Analysis:Geospatial Analysis:

```python
import folium
from folium.plugins import HeatMap

# Create a base map
base_map = folium.Map(location=[41.8781, -87.6298], zoom_start=11)

# Add HeatMap
heat_data = [[row['Latitude'], row['Longitude']] for index, row in df.iterrows()]
HeatMap(heat_data).add_to(base_map)

# Save map to an HTML file
base_map.save('crime_heatmap.html')
```

Crime Type Analysis:

```python
# Distribution of Crime Types
crime_types = df['Primary Type'].value_counts()
plt.figure(figsize=(15, 7))
crime_types.plot(kind='bar')
plt.title('Distribution of Crime Types')
plt.xlabel('Crime Type')
plt.ylabel('Number of Crimes')
plt.show()
```

Arrest and Domestic Incident Analysis:

```python
# Arrest Rates
arrest_rates = df.groupby('Primary Type')['Arrest'].mean()
plt.figure(figsize=(15, 7))
arrest_rates.plot(kind='bar')
plt.title('Arrest Rates by Crime Type')
plt.xlabel('Crime Type')
plt.ylabel('Arrest Rate')
plt.show()

# Domestic vs. Non-Domestic Crimes
domestic_crimes = df[df['Domestic'] == True]
non_domestic_crimes = df[df['Domestic'] == False]
```

```python
import pandas as pd
import numpy as np

# Sample data generation

# Crime data
np.random.seed(42)
num_records = 1000
df = pd.DataFrame({
    'Latitude': np.random.uniform(41.6, 42.0, num_records),
    'Longitude': np.random.uniform(-87.9, -87.5, num_records),
    'Crime Type': np.random.choice(['Theft', 'Assault', 'Burglary', 'Robbery', 'Vandalism'], num_records),
    'Arrest': np.random.choice([True, False], num_records),
    'Domestic': np.random.choice([True, False], num_records),
    'Date': pd.date_range(start='1/1/2020', periods=num_records, freq='H')
})

# Annual trends
annual_trends = df['Date'].dt.year.value_counts().sort_index().reset_index()
annual_trends.columns = ['Year', 'Incidents']

# Monthly trends
monthly_trends = df['Date'].dt.to_period('M').value_counts().sort_index().reset_index()
monthly_trends.columns = ['Month', 'Incidents']

# Peak hours
peak_hours = df['Date'].dt.hour.value_counts().sort_index().reset_index()
peak_hours.columns = ['Hour', 'Incidents']

# Crime types distribution
crime_types = df['Crime Type'].value_counts().reset_index()
crime_types.columns = ['Crime Type', 'Incidents']

# Arrest rates by crime type
arrest_rates = df.groupby('Crime Type')['Arrest'].mean().reset_index()
arrest_rates.columns = ['Crime Type', 'Arrest Rate']

# Domestic vs Non-Domestic
domestic_vs_non_domestic = df['Domestic'].value_counts().reset_index()
domestic_vs_non_domestic.columns = ['Domestic', 'Incidents']
domestic_vs_non_domestic['Domestic'] = domestic_vs_non_domestic['Domestic'].map({True: 'Domestic', False: 'Non-Domestic'})

# Save data to CSV files
df.to_csv('chicago_crime_data.csv', index=False)
annual_trends.to_csv('annual_trends.csv', index=False)
monthly_trends.to_csv('monthly_trends.csv', index=False)
peak_hours.to_csv('peak_hours.csv', index=False)
crime_types.to_csv('crime_types.csv', index=False)
arrest_rates.to_csv('arrest_rates.csv', index=False)
domestic_vs_non_domestic.to_csv('domestic_vs_non_domestic.csv', index=False)
```
